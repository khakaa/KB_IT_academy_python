```python
df1 = pd.DataFrame({
    '고객명': ['박세리', '이대호', '손흥민', '김연아', '마이클조던'],
    '생년월일': ['1980-01-02', '1982-02-22', '1993-06-12', '1988-10-16', '1970-03-03'],
    '성별': ['여자', '남자', '남자', '여자', '남자']})

df2 = pd.DataFrame({
    '고객명': ['김연아', '박세리', '손흥민', '이대호', '타이거우즈'],
    '연봉': ['2000원', '3000원', '1500원', '2500원', '3500원']})
```

```python
pd.merge(df1, df2)
# 마이클조던, 타이거우즈는 없어짐. 양쪽에 존재하지 않아서
```

```python
# 마이클조던, 타이거우즈를 살려보자
pd.merge(df1, df2, how='left')
'''
df1은 left, df2는 right임
how에 left를 줘서 df1은 다 살리고 df2는 공통값만 살림
그래서 마이클조던만 살고 타이거우즈는 없음
'''
```

```python
pd.merge(df1, df2, how='right') #타이거우즈만 살음
```

```python
pd.merge(df1, df2, how='outer') # 양쪽 다 살림
# 공통칼럼인 고객명의 데이터를 전부 살림
```

```python
df1 = pd.DataFrame({
    '이름': ['박세리', '이대호', '손흥민', '김연아', '마이클조던'],
    '생년월일': ['1980-01-02', '1982-02-22', '1993-06-12', '1988-10-16', '1970-03-03'],
    '성별': ['여자', '남자', '남자', '여자', '남자']})

df2 = pd.DataFrame({
    '고객명': ['김연아', '박세리', '손흥민', '이대호', '타이거우즈'],
    '연봉': ['2000원', '3000원', '1500원', '2500원', '3500원']})
```

```python
pd.merge(df1, df2, left_on="이름",
        right_on="고객명", how='outer'
        )
```

```python
ti["survived"].value_counts() # 생존자 수 
```

```python
ti.groupby("embarked")["survived"].sum() # 항구별 생존자 수
```

```python
ti.groupby("embarked")["survived"].mean() #항구별 생존율
```

```python
'''
S(사우스햄튼)에서 탑승한사람들이 젤 많이 살았는데 생존율이 젤 낮음
why?
'''
```

```python
ti.groupby("embarked")["survived"].count()
# S에서 탑승한사람이 644명인데 
# 217명이 살았지만 그거보다 더 많이 죽어서 생존율이 낮음
```

```python
ti.groupby("embarked")["survived"].agg(["count","sum","mean"])
```

```python
# 성별에 따른 탑승자수, 생존자수, 생존율
ti.groupby("sex")["survived"].agg(["count","sum","mean"]) 
```

```python
# 혼자인 경우에 따른 탑승자수, 생존자수, 생존율
ti.groupby("alone")["survived"].agg(["count","sum","mean"]) 
```

```python
# 등급에 따른 탑승자수, 생존자수, 생존율
ti.groupby("pclass")["survived"].agg(["count","sum","mean"]) 
```

```python
# who에 따른 탑승자수, 생존자수, 생존율
ti.groupby("who")["survived"].agg(["count","sum","mean"]) 
```

```python
ti.groupby(["who","pclass"])["survived"].agg(["count","sum","mean"]) 
```

```python
result = ti.groupby(["who","pclass"])["survived"].agg(["count","sum","mean"]) 
result = result.reset_index()
result.sort_values(by='mean',ascending=False).reset_index(drop=True)
```

```python
# 등급별 연령별 평균요금 비교

ti.groupby(["who","pclass"])[["fare"]].mean()
```

```python
# 요금 기준 상위 10%의 생존율 확인
rich = ti['fare'].quantile(0.9)
rich  # 결과값 : 77.9583

ti.loc[ti['fare'] >= rich, 'survived'].agg(['count','mean'])
```
