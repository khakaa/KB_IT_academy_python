```python
apt_price["시군구"].str.split()
# str을 통해서 split하면 내부적으로 스플릿을 반복실행, 
# 그 결과를 시리즈로 모아서 반복실행 시켜줌 
# 실행 결과도 시리즈
```

```python
apt_price["구"] = apt_price["시군구"].str.split().str[1] #구만 모아서 보여줌
apt_price["동"] = apt_price["시군구"].str.split().str[2] #동만 모아서 보여줌

# 전용면적을 평 변환
apt_price["평"] = apt_price["전용면적(㎡)"] / 3.03

# 거래금액 단위를 억원으로 변환
apt_price["거래금액_억원"] = apt_price["거래금액(만원)"] / 10000

# 계약년월 -> 게약년, 계약월로 분리
apt_price["계약년"] = apt_price["계약년월"] // 100 #100으로 나눠서 몫 추출
apt_price["계약월"] = apt_price["계약년월"] % 100 #100으로 나눠서 나머지 추출

apt_price.head()
```

```python
# 평정보를 이용하여 평형정보 생성
# 10평이하, 10평대, 20평대, 30평대, 40평대, 40평이상

apt_price["평형"] = pd.cut(
    apt_price["평"],
    bins=[0,10,20,30,40,50,1000],
    labels = "10평이하,10평대,20평대,30평대, 40평대, 40평이상".split(",")
)
apt_price.head()
```

```python
def pp(x):
    if x < 10 : # x는 숫자가 넘어옴
        return "아주 소형"
    elif x < 20:
        return "소형"
    else :
        return "일반"

apt_price["평"].apply( pp )

# series를 보면 s.replace 리플레이스라는 함수를 이용해서 변환하기도 함
# replace 함수는 특정 값일 경우에만 바꿀수 있음 (단점)
# s.replae({ 5:"a" }) (5=값)
# 범위를 이용한 변환을 할때는 apply를 이용한 if문을 작성하자
```

```python
# 평 정보를 평형으로 전환을 apply() 사용
apt_price["평"].describe() # 다양한 통계량을 요약

def 평형변환(x):
    if x<10:
        return "10평이하"
    if x<20:
        return "10평대"
    if x<30:
        return "20평대"
    if x<40:
        return "30평대"
    if x<50:
        return "40평대"
    if x>50:
        return "40평이상"
    
# elif로 안써도 됨. return하면 끝나니까 if로 한거임
# return말고 변수를 넣는다거나 그러면 곤란쓰. 그럴떄 elif 사용 ㄱㄱ

apt_price["평"].apply( 평형변환 )
```
