### fancy indexing
- index를 선택하여 list로 정의하고, 선택한 index list로 indexing하는 방법
- 연속적으로 추출 : 슬라이싱, 띄엄뜨엄 추출 : fancy indexing
- s[['a','c']] : fancy indexing
- s['a','c'] : 행, 열 

```python
s2 = pd.Series([70,80,90], index=[10, 10, 30]) # 인덱스 값 설정할 때 중복하지 않도록 주의해야함

s2[[10, 30]], s2[10]
```

```python
money = apt_price['거래금액(만원)']
money.index = apt_price['단지명'] # 단지명을 index로 거래금액을 value로 설정

money
# apt_price['단지명'] 아파트 이름
```


### boolean indexing
- boolean indexing은 index list에서 True인 index만 선택
- 주의해야할 점은 반드시 boolean index list의 개수와 Series의 개수가 일치해야 함
- 조건을 걸어서 list를 먼저 만들어 준 뒤 대입할 수 있음

```python
apt_price['해제사유발생일'][apt_price['해제사유발생일'].notna()] # 결측치 제외한 값
```

## pandas 데이터 선택
### loc - indexing/slicing
- [행, 열]
- slicing은 [시작(포함) : 끝(포함)]


```python
# case1. 값 대입 불가능

cond = ti['who'] == 'man'
ti[cond]['age'] = 10
```

```python
ti[cond]
```

```python
# case2. loc를 사용하면 값 대입 가능

ti.loc[ ti['who'] == 'man', 'age' ] = 10
ti
```

<br/>

### loc 다중 조건
- 다중조건은 먼저 condition을 정의하고 &와 | 연산자로 복합 조건을 생성
- & 연산자는 AND를 의미하여 조건1과 조건2가 모두 True인 경우
- | 연산자는 OR를 의미하며 조건1 또는 조건2 중 하나만 True여도 데이터 가져옴

```python
cond2 = ti['age'] >= 30
cond3 = ti['sex'] == 'male'

ti.loc[cond2 & cond3].sort_values(by='fare', ascending=False).head(10)
```
