# 중고차 판매 정보 분석 실습

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.rcParams['font.family'] = 'AppleGothic'
plt.rcParams['axes.unicode_minus'] = False
```

```python
# vehicles.csv 파일을 읽어 df 변수에 저장하세요.
df = pd.read_csv(
    './vehicles.csv',
#     encoding='cp949'
)
df
```

```python
# Unnamed: 0 컬럼을 제거하세요.
df = df.drop('Unnamed: 0', axis=1)
df
```

```python
# 가격 컬럼의 단위를 만원으로 변경하세요.

df['가격'] = df['가격'] / 10000

df
```

```python
# 제조사별 판매 차량 수를 출력하세요.
df['제조사'].value_counts()
```

```python
# 제조사별 판매 차량 수를 파이차트로 시각화하세요.
plt.figure(figsize = (10,10))

x = df['제조사'].value_counts().values
labels = df['제조사'].value_counts().index






plt.title('시장점유율')
plt.pie(x, labels=labels, autopct='%.2f%%')
plt.show()
```

```python
# 유형별 판매 차량 수를 출력하세요.
df['유형'].value_counts()
```

```python
# 유형별 판매 차량 수를 bar 차트로 시각화하세요.
plt.figure(figsize=(10, 7))

x = df['유형'].value_counts().index
y = df['유형'].value_counts()


plt.title('유형 별 판매 수')
plt.xlabel('유형')
plt.ylabel('판매 수')

# plt.bar(x,y)
sns.barplot(x,y)
plt.show()
```

```python
# 판매 가격이 가장 높은 데이터 6개를 출력하세요.
df.sort_values(by='가격', ascending=False).head(6)
```

```python
# 판매 가격이 가장 높은 6개 제품을 '제조사:모델' 대비 가격의 형태로 bar 차트 시각화하세요.
plt.figure(figsize=(10, 7))

top6 = df.sort_values(by='가격', ascending=False).head(6)
x = top6['제조사'] + ':' + top6['모델']
y = top6['가격'].values


sns.barplot(y, x)
plt.title('판매 가격이 높은 상위 6개 제품 가격')
plt.xlabel('가격 (만원)')
plt.ylabel('제조사-모델명')
plt.xlim(20000, 50000)

plt.show()
```

```python
# 평균 판매 가격이 높은 6개 제품을 '제조사:모델' 대비 가격의 형태로 bar 차트 시각화하세요.
plt.figure(figsize=(10, 7))

top6 = df.groupby(['제조사','모델'])[['가격']].mean().sort_values(by='가격', ascending=False).head(6).reset_index()
y = top6['가격'].values
x = top6['제조사'] + ':' + top6['모델']


sns.barplot(y, x)
plt.title('평균 판매 가격이 높은 상위 6개 제품 가격')
plt.xlabel('가격 (만원)')
plt.ylabel('제조사-모델명')
plt.xlim(10000, 30000)

plt.show()
```

```python
# 연식, 주행거리, 가격 간의 상관 계수를 출력하세요.
df[['연식','주행거리','가격']].corr()
```

```python
# 연식, 주행거리, 가격 간의 관계를 pairplot을 이용하여 상태 별로 시각화하세요.

sns.pairplot(df[['연식','주행거리','가격','상태']],hue='상태')

plt.show()
```

```python
# 주행거리 열을 히스토그램으로 시각화하세요.
plt.figure(figsize=(10, 7))

sns.displot(df['주행거리'])
plt.show()
```

```python
# 주행거리 열을 로그로 변환하여 주행거리_로그 열에 저장하세요. (그대로 실행하세요)
import numpy as np

df['주행거리_로그'] = np.log1p(df['주행거리'])
df.head()
```

```python
# 주행거리 로그 열을 히스토그램으로 시각화하세요.
plt.figure(figsize=(10, 7))

sns.displot(df['주행거리_로그'])
plt.show()
```

```python
# df에 새 열 '변속기_변환'을 추가하세요.
# 값은 아래 조건에 따라 설정하세요.
# 변속기의 값이 기타면 0
# 변속기의 값이 수동이면 1
# 변속기의 값이 자동이면 2

condition_list = [
    (df['변속기'] == '기타'),
    (df['변속기'] == '수동'),
    (df['변속기'] == '자동')
]
choice_list = [0,1,2]

df['변속기_변환'] = np.select(condition_list, choice_list, default='Error')
df['변속기_변환']
```

```python
# 연식, 주행거리, 가격, 변속기 간의 관계를 버블 차트(scatter)로 시각화하세요.
# x=연식, y=주행거리_로그, s=가격/30, c=변속기_변환
plt.figure(figsize=(15, 10))

plt.scatter(
    x=df['연식'],
    y=df['주행거리_로그'],
    s=df['가격'] / 30,
    c=pd.Categorical(df['변속기_변환']).codes
)
plt.xlabel('연식')
plt.xlabel('주행 거리 로그')


plt.show()
```

```python
# 주행거리가 100000보다 큰 데이터만 추출한 후 df_100k에 저장하세요.
df_100k = df[df['주행거리'] > 100000]
df_100k.head()
```

```python
# df_100k의 유형 별 평균 가격을 계산하고 내림차순 정렬하여 출력하세요.
df_100k_mean = df_100k.groupby('유형')['가격'].mean().sort_values(ascending=False)
df_100k_mean
```

```python
# 위 데이터를 bar 차트로 시각화하세요.
plt.figure(figsize=(10, 7))


x = df_100k_mean.index
y = df_100k_mean.values

sns.barplot(x, y)
plt.title('주행 거리가 10만 이상인 차량의 유형 별 평균 가격')
plt.xlabel('유형')
plt.ylabel('평균 가격')

plt.show()
```
