# 군집분석
- 주어진 데이터들을 특성에 따라 유사한 것끼리 묶음으로써 각 유형별 특징을 분석
- 같은 군집 내 데이터들의 거리를 최소화하거나 다른 군집 간의 거리를 최대화하는 군집을 형성
    - 거리 척도는 유클리드 거리가 보편적으로 많이 사용
- 사용 예시
    - 고객 분류
    - 이상치 탐지
    - 검색 엔진, 차원 축소, 준지도 학습, 이미지 분할 등
    
## 군집방법론
- 계층적
    - 가까운 개체끼리 묶거나 멀리 떨어진 개체를 차례로 분리
    - 한 번 군집에 할당된 객체는 다리 분리 x
    
- 비계층적
    - 랜덤으로 데이터를 군집화하고 군집 과정에서 값의 변화에 따라 각 데이터들을 적절한 군집으로 이동시키면서 분리
    
## K-Means
- 군집 형성 과정
    1. 랜덤하게 K개의 중앙점 지정
    2. 유클리드 거리를 이용하여 군집의 중심점과의 거리 계산 및 가장 거리가 가까운 군집으로 군집 할당
    3. 군집 내 데이터의 평균 계산 후 새로운 중심점으로 배정 및 중심점이 더 이상 이동 안할 때까지 반복

```python
from sklearn.datasets import make_blobs
plt.rc('font', family='AppleGothic')
import matplotlib.pyplot as plt

x,y = make_blobs(random_state=10)

plt.scatter(x[:,0], x[:,1])
```

```python
y
```

```python
p1 = [5.683397508810986, 5.055340128019022]
p2 = [6.68000164584778, -5.1538014055317571]
p3 = [-1.3177517986543703, -4.979922671990228]
```

```python
plt.scatter(x[:,0], x[:, 1], s=40, label='Sample')
plt.scatter(p1[0], p1[1], s=100, label='Center 1', c='orange', marker='^')
plt.scatter(p2[0], p2[1], s=100, label='Center 2', c='green', marker='^')
plt.scatter(p3[0], p3[1], s=100, label='Center 3', c='red', marker='^')
plt.legend()
plt.savefig('c1')
```

```python

# 각점에서 중심점까지의 직선거리 계산
import numpy as np
import pandas as pd
dist = pd.DataFrame(x)
dist['c1'] = np.sqrt(np.power(x[:, 0] - p1[0], 2) + np.power(x[:, 1] - p1[1], 2))
dist['c2'] = np.sqrt(np.power(x[:, 0] - p2[0], 2) + np.power(x[:, 1] - p2[1], 2))
dist['c3'] = np.sqrt(np.power(x[:, 0] - p3[0], 2) + np.power(x[:, 1] - p3[1], 2))
dist
```

```python

# 가장 가까운 중심점 찾기
dist['round1'] = dist.loc[:, 'c1':'c3'].values.argmin(axis=1)
dist
```

```python
# 가장 가까운 중심점과 같은 색으로 칠하기
color = ['orange', 'green', 'red']
for cls in range(3):
    subset = dist.loc[dist['round1']==cls, [0, 1]]
    plt.scatter(subset.iloc[:, 0], subset.iloc[:, 1], s=40, c=color[cls])
    print(subset.mean())
plt.scatter(p1[0], p1[1], s=100, label='Center 1', c='orange', marker='^')
plt.scatter(p2[0], p2[1], s=100, label='Center 2', c='green', marker='^')
plt.scatter(p3[0], p3[1], s=100, label='Center 3', c='red', marker='^')
plt.legend()
plt.savefig('c2')
```

```python
# 새로운 중심 좌표 설정
p1 = [2.840507, 4.858404]
p2 = [5.618347, -9.542885]
p3 = [0.089212,  -5.464417]
```

```python
color = ['orange', 'green', 'red']
for cls in range(3):
    subset = dist.loc[dist['round1']==cls, [0, 1]]
    plt.scatter(subset.iloc[:, 0], subset.iloc[:, 1], s=40, label='Sample')
    print(subset.mean())
plt.scatter(p1[0], p1[1], s=100, label='Center 1', c='orange', marker='^')
plt.scatter(p2[0], p2[1], s=100, label='Center 2', c='green', marker='^')
plt.scatter(p3[0], p3[1], s=100, label='Center 3', c='red', marker='^')
plt.legend()
plt.savefig('c1')
```

```python
# KMeans를 이용한 그룹찾기
# 실행할때마다 위치가 랜덤하게 달라짐

from sklearn.cluster import KMeans
color = ['orange', 'green', 'red', 'cyan', 'yellow']

# 중심값이 움직이지 않을 때까지 계산
K = KMeans(n_clusters=5).fit(x)  # n_cluster=3 3개의그룹으로 나눈다.
plt.scatter(x[:,0], x[:, 1], s=40, label='Sample')

centers = K.cluster_centers_ # 군집별 중심점 조회 가능

for idx, center in enumerate(centers):
    plt.scatter(center[0], center[1], s=100, 
                label=f'Center {idx+1}', c=color[idx], marker='^')
plt.legend()
plt.show()
```

## DBSCAN
- 밀도 기반의 클러스터링으로 점들이 몰려 있어서 밀도가 높은 부분을 군집화하는 방식
- 한 점을 기준으로 반경거리(eps) 안에 점이 N개 이상 있으면 하나의 군집으로 인식
- 군집 형성 과정
    1. 특정 점 P에서 epsilon 거리 안에 있는 점의 수 확인
    2. 군집에 할당된 점에서 epsilon 거리 안에 있는 모든 점을 같은 군집으로 할당
    3. 어떤 점을 중심으로 하여도 군집 형성 조건에 속하지 않는 점의 경우 노이즈로 분류
    
- min_samples : 그룹이냐 노이즈냐 결정하는 최소 그룹수

```python
# DBSCAN 예제

from sklearn.datasets import make_moons
x, y = make_moons(200, noise=.05, random_state=0)
plt.scatter(x[:, 0], x[:, 1], c=y, s=50)
```

```python
# KMeans로 분석 -> 원하는 대로 군집이 안됨
labels = KMeans(n_clusters=2, random_state=0).fit_predict(x)
plt.scatter(x[:,0], x[:,1], c=labels, s=50)
```

```python
# DBSCAN
from sklearn.cluster import DBSCAN
labels = DBSCAN().fit_predict(x)
plt.scatter(x[:,0], x[:,1], c=labels, s=50)
```

- 모든 데이터가 하나의 군집으로 지정되었다.
    - min_samples 또는 eps 값의 조절 필요
    - 최우선적으로 표준화 적용
    - 밀도를 기반으로 하기 때문에 데이터의 스케일에 영향이 있다.

```python
# StandardScaler를 이용해 데이터 표준화

from sklearn.preprocessing import StandardScaler
x_scaled = StandardScaler().fit_transform(x)
x_scaled.mean(axis=0), x_scaled.std(axis==0)
```

```python
labels = DBSCAN().fit_predict(x_scaled)
plt.scatter(x_scaled[:,0], x_scaled[:,1], c=labels, s=50)
```

- 표준화 이후 DBSCAN하면 반원별로 그룹핑이 잘 되었다.

# 군집 알고리즘 평가
- ARI(Adjusted Rand Index): 군집간 유사도 측정
- NMI(Normalized Mutual Infomation): 집단간 거리도
- 1 최적, 0. -1 나쁨
- 각 데이터와 가장 가까운 중심점 사이의 평균 제곱거리를 이용하는 이너셔를 사용하면, K값에 따른 이너셔의 변화를 확인, 적정 군집 수 K를 정의할 수 있음

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics.cluster import adjusted_rand_score
from sklearn.metrics.cluster import normalized_mutual_info_score

x, y = make_blobs(random_state=10)
labels = KMeans(n_clusters=3).fit_predict(x) # 3개로 나눈게 가장 좋다.

print('ACC: ', accuracy_score(y, labels))
print('ARI :', adjusted_rand_score(y, labels))
print('NMI :', normalized_mutual_info_score(y, labels))
```

```python
from sklearn.datasets import load_iris
iris = load_iris()

k_range = range(2, 15)
inertias = []

for k in k_range:
    model = KMeans(n_clusters=k).fit(iris['data'])
    inertias.append(model.inertia_)
plt.figure(figsize=(10, 5))
plt.plot(k_range, inertias)
plt.xticks(k_range)
plt.grid()

```

```python
# 실루엣 계수 계산
from sklearn.metrics import silhouette_score

k_range = range(2, 15)
scores = []

for k in k_range:
    model = KMeans(n_clusters=k).fit(iris['data'])
    scores.append(silhouette_score(iris['data'], model.labels_))


plt.figure(figsize=(10, 5))
plt.plot(k_range, scores)
plt.xticks(k_range)
plt.grid()
```
