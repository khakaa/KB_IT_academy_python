# 홈팀별 데이터 정리

```python
N = df['경기번호'].unique().shape[0] # 경기 갯수

df1 = pd.DataFrame(columns=['경기번호', '경기날짜', '팀명', '원정팀', '결과'])

for i in range(N):
    df1.loc[i, '경기번호'] = i + 1
    df1.loc[i, '경기날짜'] = df.loc[df['경기번호'] == i+1]['경기날짜'].unique()[0]
    df1.loc[i, '팀명'] = df.loc[df['경기번호'] == i+1]['팀명'].unique()[0]
    df1.loc[i, '원정팀'] = df.loc[df['경기번호'] == i+1]['팀명'].unique()[1]
    df1.loc[i, '결과'] = df.loc[df['경기번호'] == i+1]['결과'].unique()[0]
    
df1
```

# 배구 경기 관련 변수 합계 및 평균

```python
df2 = df.copy()

df2.drop(columns=['경기번호', 'No.', '이름', '결과', '출전세트_1set', '출전세트_2set', '출전세트_3set', '출전세트_4set', '출전세트_5set', 'Unnamed: 0'], inplace=True)
df_info = df2.dtypes.to_frame()
df_type_int = df_info[df_info[0] == 'int64'].index
df_type_float = df_info[df_info[0] == 'float64'].index

df2_sum = df2.groupby(['경기날짜', '팀명'])[df_type_int].sum()

# 공격_성공률, 서브_성공률, 블로킹_성공률, 리시브_성공률, 디그_성공률, 세트_성공률 팀별 평균내기

df2_sum_copy = df2_sum.copy()

df2_sum_copy['공격_성공률'] = df2_sum_copy['공격종합_성공'] / df2_sum_copy['공격종합_시도']
df2_sum_copy['오픈_성공률'] = df2_sum_copy['오픈_성공'] / df2_sum_copy['오픈_시도']
df2_sum_copy['시간차_성공률'] = df2_sum_copy['시간차_성공'] / df2_sum_copy['시간차_시도']
df2_sum_copy['이동_성공률'] = df2_sum_copy['이동_성공'] / df2_sum_copy['이동_시도']
df2_sum_copy['후위_성공률'] = df2_sum_copy['후위_성공'] / df2_sum_copy['후위_시도']
df2_sum_copy['속공_성공률'] = df2_sum_copy['속공_성공'] / df2_sum_copy['속공_시도']
df2_sum_copy['퀵오픈_성공률'] = df2_sum_copy['퀵오픈_성공'] / df2_sum_copy['퀵오픈_시도']
df2_sum_copy['서브_성공률'] = df2_sum_copy['서브_성공'] / df2_sum_copy['서브_시도']
df2_sum_copy['디그_성공률'] = df2_sum_copy['디그_성공'] / df2_sum_copy['디그_시도']
df2_sum_copy['세트_성공률'] = df2_sum_copy['세트_성공'] / df2_sum_copy['세트_시도']
df2_sum_copy['리시브_성공률'] = df2_sum_copy['리시브_정확'] / df2_sum_copy['리시브_시도']
df2_sum_copy['블로킹_성공률'] = df2_sum_copy['블로킹_성공'] / df2_sum_copy['블로킹_시도']
```

# 홈팀별 데이터와 배구 경기 관련 데이터 합치기

```python
df3 = pd.merge(left=df1, right=df2_sum_copy, on=['경기날짜', '팀명'], how='left')

df3
```

## 다중공선성 제거
- VIF가 10이 넘으면 다중공선성이 있다고 판단한다.
- 독립변수끼리 상관관계를 보이는 것을 다중공선성이라고 한다.
- 서로 다중공선성을 보이는 변수 둘 중 하나를 제거해준다.

```python
df3.fillna(df.mean(), inplace=True)

# df4 = df3.loc[:,['경기번호', '경기날짜', '결과int','팀명', '원정팀', "공격_성공률", "오픈_성공률" ,"시간차_성공률", '이동_성공률' ,'후위_성공률', '속공_성공률','퀵오픈_성공률','서브_성공률','디그_성공률','세트_성공률','리시브_성공률','블로킹_성공률', '범실_범실',]] 
# -> 공격(1894), 세트(1015) 간의 다중공선성이 높다.

df4 = df3.loc[:,['경기번호', '경기날짜', '결과int','팀명', '원정팀','공격_성공률',"시간차_성공률", '이동_성공률' ,'후위_성공률', '속공_성공률','서브_성공률','리시브_성공률','블로킹_성공률', '범실_범실']] 

```

```python
df5 = df4.drop(['경기번호', '경기날짜', '결과int','팀명', '원정팀'],axis='columns')

from statsmodels.stats.outliers_influence import variance_inflation_factor
def vif(x):
    vif = pd.DataFrame()
    vif['VIF_Factor'] = [variance_inflation_factor(x.values,i)
         for i in range(x.shape[1])]
    vif['feature'] = x.columns
    return vif

vif(df5).sort_values(by='VIF_Factor',ascending=False)
```

## 선형회귀분석
- R-squared (결정계수)
    - 모형이 데이터에 잘 맞는 정도를 보여주는 지표 중 하나 , 1에 가까울수록 데이터를 잘 설명해주는 상태
    - 보통 0.4이상이면 괜찮은 모델이라고 할 수 있다.

```python
import statsmodels.api as sm

x = df5
y = df4['결과int']

results = sm.OLS(y, sm.add_constant(x)).fit()
print(results.summary())
```
