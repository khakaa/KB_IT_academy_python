```python
import pandas as pd
df = pd.read_excel('student_info.xls')
df
```

## 수업별 군집분석

```python
df2 = df.copy()

# Mark, Attended 단위가 다르기 때문에 정규화해준다.
from sklearn.preprocessing import MinMaxScaler
df2[['Mark', 'Attended']] = MinMaxScaler().fit_transform(df2[['Mark', 'Attended']])
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

plt.figure(figsize=(10,6))

for k in range(2,8):
    model = KMeans(n_clusters=k)
    pred = model.fit_predict(df2[['Attended', 'Mark']])
    

    plt.subplot(2,3,k-1)     # 2행 3열 그래프
    plt.tight_layout() # 그래프들끼리 겹치지 않고 적당한 간격을 둠
    
    plt.title(f'k={k}')
    plt.xlabel('Attended')
    plt.ylabel('Mark')

    plt.scatter(
        df2['Attended'], # x 
        df2['Mark'], # y
        c=pred # color : 같은 값끼리 같은 색을 가짐
        )
```

```python
model = KMeans(n_clusters=4)
pred = model.fit_predict(df2[['Mark', 'Attended']])
plt.figure(figsize=(20,6))

plt.scatter(
    df['Mark'], # x 
    df['Attended'], # y
    c=pred
    )


# 각 데이터에 제품번호 출력
for index, semester, name, course, mark, attended in df.itertuples():
    # 주석 달기
    plt.annotate(f'{pred[index]} : {course}', (mark+0.1, attended+0.1), fontsize=14) # +0.5하면 점에 글자가 겹쳐지지 않음
plt.show()
```

## 학생별 군집분석

```python
df3 = df.groupby('Name')[['Mark','Attended']].mean()
df3
```

```python
plt.figure(figsize=(10,6))

for k in range(2,8):
    model = KMeans(n_clusters=k)
    pred = model.fit_predict(df3[['Attended', 'Mark']])
    

    plt.subplot(2,3,k-1)     # 2행 3열 그래프
    plt.tight_layout() # 그래프들끼리 겹치지 않고 적당한 간격을 둠
    
    plt.title(f'k={k}')
    plt.xlabel('Attended')
    plt.ylabel('Mark')

    plt.scatter(
        df3['Attended'], # x 
        df3['Mark'], # y
        c=pred # color : 같은 값끼리 같은 색을 가짐
        )
```

```python
model = KMeans(n_clusters=4)
pred = model.fit_predict(df3[['Attended', 'Mark']])
# plt.figure(figsize=(20,6))

plt.scatter(
    df3['Attended'], # x 
    df3['Mark'], # y
    c=pred
    )


# 각 데이터에 제품번호 출력
for name, mark, attended in df3.itertuples():
    # 주석 달기
    plt.annotate(f'{name}', ( attended+0.1, mark+0.1), fontsize=14) # +0.5하면 점에 글자가 겹쳐지지 않음
plt.show()
```
