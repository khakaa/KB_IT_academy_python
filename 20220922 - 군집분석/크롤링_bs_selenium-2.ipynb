## urllib.parse.urlencode()함수를 이용한 매개변수의 URL을 인코딩
- http://www.kma.go.kr/weather/forecast/mid-term-rss3.jsp
    - stnId: 지역지정
    - 전국 : 108, 서울경기:109, 강원도:105, 충청북도: 131, 충청남도: 133, 전라북도: 146, 전라남도: 156, 경상북도: 143, 경상남도:159, 제주특별자치도: 184

````python
import urllib.request
import urllib.parse
api="http://www.kma.go.kr/weather/forecast/mid-term-rss3.jsp"
values={'stdid':108}
params=urllib.parse.urlencode(values)
url=api+'?'+params
print(url)
data=urllib.request.urlopen(url).read()
text=data.decode("utf-8")
print(text)
````

## BeautifulSoup을 이용한 데이터 수집
- BeautifulSoup: 웹크롤링을 도와주는 라이브러리. HTML,XML에서 데이터를 추출 가능
- pip install beautifulsoup4
- 문서파싱하기  
    - soup=BeautifulSoup(res,"html.parser")
- 찾기1: 첫번째 태그를 찾기
    - soup.title.text : 문서에서 첫번째 title태그의 글자정보 추출
    - soup.a.text : 문서에서 첫번째 a태그의 글자정보 추출
    - .(점) 이용할 경우 첫번째 태그 추출
- 찾기2 : 속성 찾기
    - soup.title.attrs -> 모든 속성 출력
    - soup.title["href"] -> title의 href 속성을 출력
- 찾기3 : find()함수 / find_all() 함수 -> 태그이름과 속성을 조건으로 검색
    - a = soup.find('a', attrs={"class": "Nbtn_upload"}) : a태그중 class속성값이 Nbtn_upload인 첫번째 태그 검색  
    - a_list = soup.find_all('a', attrs={"class": "Nbtn_upload"}) : a태그중 class속성값이 Nbtn_upload인 모든 태그 검색 후 리스트 반환
- 찾기4 : select()함수 / select_all()함수 -> css 선택자를 이용하여 원하는 태그 검색. 가장 다양한 형태로 검색 가능
    - el = soup.select("a.Nbtn_upload")  : a태그중 class속성값이 Nbtn_upload인 첫번째 태그 검색  
    - el_list = soup.select_all("a.Nbtn_upload")  : a태그중 class속성값이 Nbtn_upload인 모든 태그 검색 후 리스트 반환

- pip install requests

GET 방식: requests.get()  
POST 방식: requests.post()

DELETE 방식: requests.delete()
- 응답코드: status_code
```python
>>> import requests

```
- Response 객체
```python
    - response.request : 내가 보낸 request 객체에 접근 가능    
    - response.status_code:응답 코드
    - response.raise_for_status(): 200 (OK 코드)이 아닌 경우 에러 raise    
    - response.json(): json response일 경우 딕셔너리 타입으로 바로 변환
    - response.content: content 속성을 통해 바이너리 타입으로 데이터를 받을 수 있다.  
    - response.text: text 속성을 통해 UTF-8로 인코딩된 문자열을 받을 수 있다.    
    - response.encoding: encoding 정보 확인
```

### XML 파싱
- XML도 HTML과 동일한 방식으로 파싱한다
- 서울특별시 강남구 역삼1동 기상청 날씨 예보 파싱하기

````python
import requests
from bs4 import BeautifulSoup
url="https://www.kma.go.kr/wid/queryDFSRSS.jsp?zone=1168064000" #서울특별시 강남구 역삼1동
res=response = requests.get(url)
soup=BeautifulSoup(res.text,"html.parser")
# 예보 대상 동네 이름은 <category>에 저장되어 있다
print("예보동네이름:", soup.select_one("category").string )
# 예보데이터는 <data>내에 여러 태그로 싸여있다
data=soup.select("data")
for d in  data:
    hour = d.select_one("hour").string
    day = d.select_one("day").string # 0 오늘, 1 내일, 2 모레. -> if안쓰고 0,1,2를 오늘,내일,모레로 변환해보기
    temp = d.select_one("temp").string
    wfkor = d.select_one("wfkor").string
    print(hour, day,temp,wfkor)
````

#### 네이버 검색
- el = driver.find_element(By.CSS_SELECTOR, "#query")
- els = driver.find_elements(By.CSS_SELECTOR, "#query")
    - By.ID - 태그의 ID 값
    - By.NAME - 태그의 name값
    - By.XPATH - 태그의 경로
    - By.TAG_NAME - 태그명
    - By.CLASS_NAME - 태그의 class 값
    - By.CSS_SELECTOR - CSS 선택자

- element.send_keys( Keys )

````python
from selenium.webdriver.common.by import By
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time

driver = webdriver.Chrome('./chromedriver') #또는 chromedriver.exe
driver.implicitly_wait(15) # 묵시적 대기, 활성화를 최대 15초가지 기다린다.

# 페이지 가져오기(이동)
driver.get('https://www.naver.com')

# 요소 찾기 - 검색창찾고 키 전송
search = driver.find_element(By.CSS_SELECTOR, "#query") # 셀레니움 4로 바뀌면서 태그 검색방법이 바뀌었다. 나머지는 동일하다.
# search = driver.find_element_by_css_selector('#query')
search.send_keys('파이썬') # 요소에 "파이썬" 입력
search.send_keys(Keys.ENTER) # 요소에 엔터키 입력
time.sleep(2)

# 요소 찾기 - 지식백과에서 파이썬 클릭
posts = driver.find_elements(By.CSS_SELECTOR,'a.link_tit')
posts[0].click()
time.sleep(2)
````
